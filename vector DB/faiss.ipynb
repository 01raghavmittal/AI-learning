```python
import numpy as np
import faiss


d = 5  # dimension


nb = 1000
xb = np.random.random((nb, d)).astype('float32')
print(xb.shape)


nq = 1
xq = np.random.random((nq, d)).astype('float32')




index = faiss.IndexFlatL2(d)
print(index)

print(f"Is index trained? {index.is_trained}")
index.add(xb)
print(f"Total vectors in index: {index.ntotal}")

k = 5 
D, I = index.search(xq, k) 

print(f"\nDistances (D):  {D}")
print(f"Indices of nearest neighbors (I): {I[0]}")


 
nearest_neighbor_index = I[0][0]
print(f"Nearest neighbor text: {nearest_neighbor_index}")

print(f"Nearest neighbor vector: {xb[nearest_neighbor_index]}")



 
```
