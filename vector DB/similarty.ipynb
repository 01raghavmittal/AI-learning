Similarty technique

1. jaccard 


```python
def jaccard(x ,y):
    x= set(x)
    y= set(y)

    shared = x.intersection(y)
    union= x.union(y)

    return len(shared)/len(union)
    
```

```python
jaccard((1,2,3),(2,4))
```
